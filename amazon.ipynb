Data Cleaning

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
missing_images = data['img_link'].isnull().sum()
missing_links = data['product_link'].isnull().sum()

print("Missing Image Links:", missing_images)
print("Missing Product Links:", missing_links)

print("Unique image links:", data['img_link'].nunique())
print("Unique product links:", data['product_link'].nunique())
```

```python
data['discounted_price']=data['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
data['actual_price']=data['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
data['discount_percentage']=data['discount_percentage'].str.replace('%','').astype(float)
```

```python
data['rating_count']=data['rating_count'].str.replace(',','').astype(float)
```

```python
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
```

```python
data.head()
```

```python
data['product_name'] = data['product_name'].str.strip().str.title()
```

```python
data = data.groupby('product_id').agg({
    'product_name': 'first',
    'category': 'first',
    'discounted_price': 'mean',
    'actual_price': 'mean',
    'discount_percentage': 'mean',
    'rating': 'mean',
    'rating_count': 'mean',
    'about_product': 'first',
    'user_id': 'first',
    'user_name': 'first',
    'review_id': 'first',
    'review_title': 'first',
    'review_content': 'first',
    'img_link': 'first',
    'product_link': 'first',
}).reset_index()
```

```python
data.head()
```

Visualization

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['rating'], bins=20)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
data['main_category'] = data['category'].str.split('|').str[0]
plt.figure(figsize=(12, 6))
sns.boxplot(x='main_category', y='rating', data=data)
plt.title('Average Rating by Main Category')
plt.xticks(rotation=45)
plt.xlabel('Main Category')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['discounted_price'], bins=30,color='blue', label='Discounted Price')
sns.histplot(data['actual_price'], bins=30, color='orange', label='Actual Price')
plt.title('Distribution of Discounted and Actual Prices')
plt.xlabel('Price (₹)')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_brands = data['product_name'].str.split().str[0].value_counts().head(10)
sns.barplot(x=top_brands.values, y=top_brands.index)
plt.title('Top 10 Brands by Product Count')
plt.xlabel('Number of Products')
plt.ylabel('Brand')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='main_category', y='discount_percentage', data=data)
plt.title('Discount Percentage Distribution by Main Category')
plt.xticks(rotation=45)
plt.xlabel('Main Category')
plt.ylabel('Discount Percentage (%)')
plt.show()
```

Feature Engineering

```python
data['price_difference'] = data['actual_price'] - data['discounted_price']
print(data['price_difference'])
```

```python
data['rating_count_bin'] = pd.qcut(data['rating_count'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
print(data['rating_count_bin'])
```

```python
data['brand'] = data['product_name'].str.split().str[0]
print(data['brand'])
```

```python
data['review'] = np.where(data['rating'] >= 3, 'Positive', 'Negative')
print(data['review'])
```

```python
data['price_per_rating'] = data['discounted_price'] / data['rating']
data['price_per_rating'] = data['price_per_rating'].apply(lambda x:f"{x:.2f}")
```

```python
print(data['price_per_rating'])
```

```python
data['norm_discunted_price'] = data['discounted_price']-data['discounted_price'].min()/data['discounted_price'].max()-data['discounted_price'].min()
print(data['norm_discunted_price'])
```

```python
data['norm_actual_price'] = data['actual_price']-data['actual_price'].min()/data['actual_price'].max()-data['actual_price'].min()
data['norm_actual_price']
```

```python
top_brands = data['brand'].value_counts().head(10).index
print(top_brands)
```

```python
data['brand'] = data['brand'].apply(lambda x: x if x in top_brands else 'Other')
data = pd.get_dummies(data, columns=['main_category', 'brand'], prefix=['cat', 'brand'])
```
